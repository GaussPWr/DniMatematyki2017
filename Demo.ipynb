Stwórzmy pusty obrazek :)

```python
p = figure(plot_width=400, plot_height=400)
# później będziemy w tym miejscu tworzyli faktyczny wykres :) 
show(p)
```

```python
import math
import numpy as np
x = np.linspace(0,10, 100)
y1 = [math.sin(item) for item in x]
y2 = [math.cos(item) for item in x]
```

### Zwizualizujmy pierwsze dane!

```python
p = figure(plot_width=400, plot_height=400)
p.circle(x,y1) # tworzymy kółka
show(p)
```

### Czas zacząć upiekszać wykres!

```python
p = figure(plot_width=400, plot_height=400)
p.circle(x,y1, fill_color='orange') # Dodajemy kolor
show(p)
```

Jest brzydko - zmieniliśmy wypełnienie, natomiast punkty posiadają jeszcze obramowanie

Pełna lista kolorów do znalezienia np. na stronie:
* https://www.w3schools.com/cssref/css_colors.asp

```python
p = figure(plot_width=400, plot_height=400)
p.circle(x,y1, fill_color='orange', line_color=None) # Usuwamy obramowanie punktu (możemy dać inny kolor oczywiście)
show(p)
```

### Ustalmy stałą wielkość kropek

```python
p = figure(plot_width=400, plot_height=400)
p.circle(x,y1, fill_color='orange', line_color=None, size=5)  #
show(p)
```

### Oraz przejrzystość

```python
p = figure(plot_width=400, plot_height=400)
p.circle(x,y1, fill_color='orange', line_color=None, size=5, alpha=0.6)  #
show(p)
```

### Dodajmy ramkę

```python
p = figure(plot_width=400, plot_height=400)
p.circle(x,y1, fill_color='orange', line_color=None, size=5, alpha=0.6) 
p.outline_line_width = 2 #
p.outline_line_alpha = 0.8 #
p.outline_line_color = "black" #
show(p)
```

### Opatrzmy osie podpisami

```python
p = figure(plot_width=400, plot_height=400, title='Nasz pierwszy wykres') # 
p.circle(x,y1, fill_color='orange', line_color=None, size=5, alpha=0.6) # Usuwamy obramowanie punktu (możemy dać inny kolor oczywiście)

p.outline_line_width = 2
p.outline_line_alpha = 0.8
p.outline_line_color = "black"

p.xaxis.axis_label = 'Oś X' #
p.yaxis.axis_label = 'Oś Y' # 
show(p)
```

### Dopracujmy podpisy

```python
p = figure(plot_width=400, plot_height=400, title='Nasz pierwszy wykres')
p.circle(x,y1, fill_color='orange', line_color=None, size=5, alpha=0.6) 
p.outline_line_width = 2
p.outline_line_alpha = 0.8
p.outline_line_color = "black"

p.title.align = 'center' #
p.title.text_font_size = '14pt' #

p.xaxis.axis_label = 'Oś X'
p.xaxis.axis_label_text_color = 'Grey'  #
p.xaxis.axis_label_text_font_size = '10pt' #
p.yaxis.axis_label = 'Oś Y'
p.yaxis.axis_label_text_color = 'Grey' # 
p.yaxis.axis_label_text_font_size = '10pt' #

show(p)
```

### Legendy!

```python
p = figure(plot_width=400, plot_height=400, title='Nasz pierwszy wykres')
p.circle(x,y1, fill_color='orange', line_color=None, size=5, alpha=0.6, legend='sin(x)') #

p.outline_line_width = 2
p.outline_line_alpha = 0.8
p.outline_line_color = "black"

p.title.align = 'center'
p.title.text_font_size = '14pt'

p.xaxis.axis_label = 'Oś X'
p.xaxis.axis_label_text_color = 'Grey'
p.xaxis.axis_label_text_font_size = '10pt'
p.yaxis.axis_label = 'Oś Y'
p.yaxis.axis_label_text_color = 'Grey'
p.yaxis.axis_label_text_font_size = '10pt'

show(p)
```

### Legendy także możemy modyfikować

```python
p = figure(plot_width=400, plot_height=400, title='Nasz pierwszy wykres')
p.circle(x,y1, fill_color='orange', line_color=None, size=5, alpha=0.6, legend='sin(x)') 

p.outline_line_width = 2
p.outline_line_alpha = 0.8
p.outline_line_color = "black"

p.title.align = 'center'
p.title.text_font_size = '14pt'

p.xaxis.axis_label = 'Oś X'
p.xaxis.axis_label_text_color = 'Grey'
p.xaxis.axis_label_text_font_size = '10pt'
p.yaxis.axis_label = 'Oś Y'
p.yaxis.axis_label_text_color = 'Grey'
p.yaxis.axis_label_text_font_size = '10pt'

p.legend.location = 'bottom_left' # 
p.legend.border_line_width = 1 #
p.legend.border_line_color = "Grey" #
p.legend.border_line_alpha = 0.3 #
p.legend.label_text_font_style = 'italic' #
show(p)
```

## Czas na dodanie więcej danych!

```python
p = figure(plot_width=400, plot_height=400, title='Nasz pierwszy wykres')
p.circle(x, y1, fill_color='orange', line_color=None, size=5, alpha=0.6, legend='sin(x)') 
p.circle(x, y2, fill_color='blue', line_color=None, size=5, alpha=0.6, legend='cos(x)') #

p.outline_line_width = 2
p.outline_line_alpha = 0.8
p.outline_line_color = "black"

p.title.align = 'center'
p.title.text_font_size = '14pt'

p.xaxis.axis_label = 'Oś X'
p.xaxis.axis_label_text_color = 'Grey'
p.xaxis.axis_label_text_font_size = '10pt'
p.yaxis.axis_label = 'Oś Y'
p.yaxis.axis_label_text_color = 'Grey'
p.yaxis.axis_label_text_font_size = '10pt'

p.legend.location = 'bottom_left' 
p.legend.border_line_width = 1 
p.legend.border_line_color = "Grey" 
p.legend.border_line_alpha = 0.3 
p.legend.label_text_font_style = 'italic'
show(p)
```

### Zmieńmy sposób wyświetlania danych

```python
p = figure(plot_width=400, plot_height=400, title='Nasz wykres robi się coraz piękniejszy')
p.line(x, y1, line_color='orange', line_width=3, alpha=0.6, legend='sin(x)') #
p.circle(x, y1, fill_color='white', line_color='orange', size=5, alpha=0.6, legend='sin(x)') #

p.line(x, y2, line_color='blue', line_width=3, alpha=0.6, legend='cos(x)') #
p.square(x, y2, fill_color='white', line_color='blue', size=5, alpha=0.6, legend='cos(x)') #

p.outline_line_width = 2
p.outline_line_alpha = 0.8
p.outline_line_color = "black"

p.title.align = 'center'
p.title.text_font_size = '14pt'

p.xaxis.axis_label = 'Oś X'
p.xaxis.axis_label_text_color = 'Grey'
p.xaxis.axis_label_text_font_size = '10pt'
p.yaxis.axis_label = 'Oś Y'
p.yaxis.axis_label_text_color = 'Grey'
p.yaxis.axis_label_text_font_size = '10pt'

p.legend.location = 'bottom_left' 
p.legend.border_line_width = 1 
p.legend.border_line_color = "Grey" 
p.legend.border_line_alpha = 0.3 
p.legend.label_text_font_style = 'italic'
show(p)
```

### Zawsze dobrze dodać wiedzę na temat natury samych danych

```python
from bokeh.models.annotations import Span, BoxAnnotation #

p = figure(plot_width=400, plot_height=400, title='Dużo wiedzy jednocześnie')
p.line(x, y1, line_color='orange', line_width=3, alpha=0.6, legend='sin(x)') 
p.circle(x, y1, fill_color='white', line_color='orange', size=5, alpha=0.6, legend='sin(x)') 

p.line(x, y2, line_color='blue', line_width=3, alpha=0.6, legend='cos(x)') 
p.square(x, y2, fill_color='white', line_color='blue', size=5, alpha=0.6, legend='cos(x)') 

p.outline_line_width = 2
p.outline_line_alpha = 0.8
p.outline_line_color = "black"

p.title.align = 'center'
p.title.text_font_size = '14pt'

p.xaxis.axis_label = 'Oś X'
p.xaxis.axis_label_text_color = 'Grey'
p.xaxis.axis_label_text_font_size = '10pt'
p.yaxis.axis_label = 'Oś Y'
p.yaxis.axis_label_text_color = 'Grey'
p.yaxis.axis_label_text_font_size = '10pt'

p.legend.location = 'bottom_left' 
p.legend.border_line_width = 1 
p.legend.border_line_color = "Grey" 
p.legend.border_line_alpha = 0.3 
p.legend.label_text_font_style = 'italic'

mean = Span(location=0, dimension='width', line_color='grey', line_dash=[5,5], line_width=1) #
p.add_layout(mean) #

lower = BoxAnnotation(top=-1, fill_alpha=0.1, fill_color='pink') #
p.add_layout(lower) #

upper = BoxAnnotation(bottom=1, fill_alpha=0.1, fill_color='pink') #
p.add_layout(upper) #

show(p)
```

# Czas zacząć magię :)

### Interakcja

```python
x = list(range(-5, 5))
y = [item**2 for item in x]
```

```python
p = figure(plot_width=400, plot_height=400, tools='tap', title='Dotknij mnie')
p.circle(x,y, size = 10, fill_color = 'orange', line_color='navy', line_dash = [3,3],
        
        selection_fill_color = 'red',
        selection_fill_alpha = 1.,
        
        nonselection_fill_color = 'grey',
        nonselection_fill_alpha = 0.2)
show(p)
```

## Hover

```python
x = np.linspace(-2*math.pi,2*math.pi,math.ceil(4*math.pi*50))
y = [math.sin(item) for item in x]
```

```python
from bokeh.models.tools import HoverTool
p = figure(plot_width=400, plot_height=400)
p.line(x,y, line_width = 3, line_color = 'Lavender', line_dash = [7,7])
cr = p.circle(x, y, size=20,
              fill_color="grey", hover_fill_color="Blue",
              fill_alpha=0.05, hover_alpha=0.3,
              line_color=None, hover_line_color=None)
p.add_tools(HoverTool(tooltips=None, renderers=[cr], mode='hline'))

show(p)
```

# O Dataframach słów kilka 
**Bokeh** pracować może zarówno z listami, listami z *numpy* jak i z obiektami *pandas.DataFrame*.
Ta ostatnia metoda pozwala na tworzenie wykresów z tabeli jeszcze łatwiej niż robiliśmy to do tej pory!

```python
import pandas as pd
import seaborn.apionly as sns # w tej biblitece mamy różne zbiory danych
```

```python
iris = sns.load_dataset('iris') # ładujemy dane
print(iris.head()) # printujemy część zbioru danych, żeby wiedzieć jak wygląda
```

```python
p = figure(plot_width = 400, plot_height = 400)
p.circle('sepal_length', 'sepal_width', source=iris) # parametr source określa z jakich danych korzystamy
show(p)
```

```python
## Są też inne typy wykresów
from bokeh.charts import Histogram

p = Histogram(iris, 'sepal_length', title="Pierwszy histogram")

show(p)
```

### I trochę bardziej zaawansowany..

```python
## Są też inne typy wykresów
from bokeh.charts import Histogram

p = Histogram(iris, 'sepal_length', color='species', 
              legend='top_right', title="Histogram długości od gatunku")

p.xaxis.axis_label = 'Długość'
p.yaxis.axis_label = 'Ilość próbek'

show(p)
```
